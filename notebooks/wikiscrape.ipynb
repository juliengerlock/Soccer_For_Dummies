```python
import requests, os, datetime, json
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
website_url = 'https://en.wikipedia.org/wiki/2019_Major_League_Soccer_season'
response = requests.get(website_url)

response.status_code
```

```python
soup = BeautifulSoup(response.content, 'lxml')
print(soup.title)
```

```python
tables = soup.find_all('table', attrs = {'class': 'wikitable'})
tables
```

```python
result_list = pd.read_html(str(tables)) # a list of dataframes
len(result_list)
```

```python
world_soups = result_list[6]  # get the first df from the list
world_soups.head(20)
```
